```python
d = 256# no of extended elements found in an ascii table
  
# x-substring we are searching for
# y-text we are searching in
# k-lenght of substring
import string
def rh_get_match(x, y, k):
    """
    Finds all common length-k substrings of x and y
    using rolling hashing on both strings.
    Input:
    - x:pattern string
    - y: text string
    - k: int, length of substring
    Output:
    - A list of tuples (i, j) where x[i:i+k] = y[j:j+k]
    """
    #your code here
    #remove white space and punctuations
    y = y.translate(str.maketrans('', '', string.punctuation))
    x = x.translate(str.maketrans('', '', string.punctuation))
    y=y.replace(" ", "").lower()
    x=x.replace(" ", "").lower()
    N = len(x)
    p = 0    # hash values for substring
    t = 0    # hash value for text
    h = 1    # initilaizaion of our higher order 
    hashtable=[[] for _ in range(971)]
    solution=[]
    substring=[]
    for i in range(k-1):
        h = (h * d)% 971
    # Calculating the hash value of pattern and first window
    # of text
    for i in range(k):
        p = (d * p + ord(y[i]))% 971
        t = (d * t + ord(x[i]))% 971
        i=+1
    hashtable[t].append(x[:k])
    for j in range(N-k):
        if j < N-k:
            j+=1
            t = (d*(t-ord(x[j-1])*h) + ord(x[ j-1+k]))% 971
            hashtable[t].append(x[j:j+k])    
        if p==t:
            substring.append(x[j:j+k])
            solution.append((j,0))
    return solution
        
        # The rolling hash method i.e calculating hash value for 
        # next window of text:substracting the initial's value hash 
        # and adding the last digit's hash
    
# Driver program to test the above function
x='today learn french germany and italian french'
y='french'
x=x.replace(" ", "")
y=y.replace(" ", "")
k=len(y)
rh_get_match(x, y, k)
 
```

I used the multiplication hashing technique method h(k) = floor (m * frac (k * c)).The multiplication hashing technique has the advantage that it is not dependent on the table size unlike the division method.However,I have also used a modulo to prevent overun i.e having indexes larger than our table indexes.I have used arrays to store our hash items and keys since the time complexity of accessing out items is costant 0(1).However,there is an exception in that for chained lists the search time for such will be to the O(N),N being the lenght of the chained array.Even though the table size does not affect our hash function i have chosen a large table size and an odd modulo size to ensure uniform distribution thus avoiding collisions in our hash table.


```python
import math
import string

def regular_get_match(x,y,k):
    """
    Finds all common length k substrings of x and y
    not using rolling hashing on both strings.
    
    Input:
        -x,y: strings
        -k: int,length of substring
        
    Output:
        - A list of tuples(i,j)
         where x[i:i+k]= y[j:j+k]
    """
    #remove punctuation and conert everything to lower case
    y = y.replace(" ", "")
    x = x.replace(" ", "")
    y = y.translate(str.maketrans('', '', string.punctuation))
    x = x.translate(str.maketrans('', '', string.punctuation))
    

    
    
   
    M=len(x) #length of string input 1
    Q=len(y) #length of string input 2
    
    
    # store matched substrings
    match=[] 
    
    
    #array to store the hashvalues and items
    hashtable=[[] for _ in range(1971)] 
    
    # length of the hashtable that stores x-values
  
    #going through string x
    for j in range(M-k+1):
        
        #for the current substring in x
        hash_text = hash_function(x[j:j+k]) % 971
        
        #add this value to my hashtable
        hashtable[hash_text].append(x[j:j+k])
        
        # a new list where we store repeated indices
        similar=[]
        
        #now look through string y
        for w in range (Q-k+1):
            
            
            #value for the current substring in y
            hash_pattern=hash_function(y[w:w+k ]) % 971
            
            #looks through the sublists
            if hashtable[hash_pattern]:

                for sub_x in hashtable[hash_pattern]:

                    if sub_x == y[w:w+k] and x[j:j+k] == y[w:w+k]:

                        similar.append(w)
                        
                        # we break to prevent copies
                        break
                        
        for sim in similar:
            # we add the values on this list to our solution list
            match.append((j,sim))
    return match


#my hash function
def hash_function(window):
    m=len(x)
    for i in range(len(window)):
        key= math.floor(m* ord(window[i])*1)
    
    return key


x='today learn french germany and italian'
y='french'
k=len(y)

regular_get_match(x,y,k)
```
